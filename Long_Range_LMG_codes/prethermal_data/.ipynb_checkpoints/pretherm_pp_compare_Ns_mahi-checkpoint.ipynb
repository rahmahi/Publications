# Long time exact evolution

```python
from qutip import mesolve, basis, jmat, Options, expect
from multiprocessing import Pool
from tqdm import tqdm
from scipy.special import jn_zeros
import numpy as np
import matplotlib.pyplot as plt

def drive(t, args):
    w = args['omega']
    h = args['h']
    h0 = args['h0']
    return h * np.sin(w*t) + h0

def get_hamiltonians(N):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    kn =  2.0/(N-1)                                     
    H0 = kn * sz **2 
    H1 = 2 * sx
    return H0,H1

def sys_evolve(w,N,opts):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    T = 2 * np.pi/w
    h = jn_zeros(0,5)[0]* w / 4.0
    params = {'h0':np.pi/32, 'h':h,'omega':w, 'N':N} 
    H0, H1 = get_hamiltonians(N)
    H = [H0,[H1,drive]]
    en, sts = sx.eigenstates() 
    rho0 = sts[0]
    times = np.linspace(0,75*T, 750)
    #times = np.linspace(0,1000*T, 20000)
    hbar = []
    out = mesolve(H, rho0, times, [], [], args = params, options=opts)
    psi_ts = out.states
    
    for i,t in enumerate(times):
        psit = np.array(psi_ts[i].full())[:,0].reshape(2*N+1,1)
        Ht = H0 + H1 * drive(t, params)
        Ht = np.array(Ht.full())
        hbar.append(np.conjugate(psit).T @ Ht @ psit)
    hbar = (np.array(hbar)/N).reshape(len(times),1)
    return hbar.real, (times/(2*np.pi/w)).reshape(len(times),1)

if __name__ == '__main__':
    nprocs = 2
    N = 50
    omegas = [0.5210526315789473,1.5,1.9795918367346939, 2.5, 7.0, 10.0]
    p = Pool(processes = nprocs)
    print("running for TSS spin N=",N, 'nprocs=',nprocs," !")
    opts = Options(nsteps=1e5, num_cpus=nprocs, openmp_threads=1, atol=1e-12, rtol=1e-14)
    data = np.array(p.starmap(sys_evolve,tqdm([(w,N, opts) for w in omegas])))
    
    print('run complete!')
```

# Plot <<H(t)>> and S_x side by side

```python
%%time
from qutip import mesolve, basis, jmat, Options, expect, Qobj
from multiprocessing import Pool
from tqdm import tqdm
from scipy.special import jn_zeros
import numpy as np
import matplotlib.pyplot as plt

def drive(t, args):
    w = args['omega']
    h = args['h']
    h0 = args['h0']
    return h * np.sin(w*t) + h0

def get_hamiltonians(N):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    kn =  2.0/(N-1)                                     
    H0 = kn * sz **2 
    H1 = 2 * sx
    return H0,H1

def sys_evolve(w,N,opts):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    T = 2 * np.pi/w
    h = jn_zeros(0,5)[1]* w / 4.0
    params = {'h0':np.pi/32, 'h':h,'omega':w, 'N':N} 
    H0, H1 = get_hamiltonians(N)
    H = [H0,[H1,drive]]
    en, sts = sx.eigenstates() 
    rho0 = sts[0]
    times = np.linspace(0,75*T, 750)
    #times = np.linspace(0,1000*T, 20000)
    hbar , mx = [], []
    out = mesolve(H, rho0, times, [], [], args = params, options=opts)
    psi_ts = out.states
    
    for i,t in enumerate(times):
        psit = np.array(psi_ts[i].full())[:,0].reshape(2*N+1,1)
        Ht = H0 + H1 * drive(t, params)
        Ht = np.array(Ht.full())
        hbar.append(np.conjugate(psit).T @ Ht @ psit)
        mx.append(expect(sx,Qobj(psit)))
        
    hbar = (np.array(hbar)/N).reshape(len(times),1)
    mx = (np.array(mx)).reshape(len(times),1)
    
    #return hbar.real, (times/(2*np.pi/w)).reshape(len(times),1)
    return hbar.real, (times/(2*np.pi/w)).reshape(len(times),1), mx

if __name__ == '__main__':
    nprocs = 2
    N = 50
    omegas = [0.5210526315789473,1.9795918367346939, 7.0, 12.0]
    p = Pool(processes = nprocs)
    print("running for TSS spin N=",N, 'nprocs=',nprocs," !")
    opts = Options(nsteps=1e5, num_cpus=nprocs, openmp_threads=1, atol=1e-12, rtol=1e-14)
    data = np.array(p.starmap(sys_evolve,tqdm([(w,N, opts) for w in omegas])))
    
    print('run complete!')
```

# Long time average of <H(t)>

```python
%%time

from qutip import mesolve, basis, jmat, Options, expect, Qobj
from multiprocessing import Pool
from tqdm import tqdm
from scipy.special import jn_zeros
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({"figure.figsize": (12,8),"text.usetex": True,
    "font.family": "sans-serif","font.size": 15,"font.sans-serif": ["Computer Modern"]})

def drive(t, args):
    w = args['omega']
    h = args['h']
    h0 = args['h0']
    return h * np.sin(w*t) + h0

def get_hamiltonians(N):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    kn =  2.0/(N-1)                                     
    H0 = kn * sz **2 
    H1 = 2 * sx
    return H0,H1

def sys_evolve(w,N,opts):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    T = 2 * np.pi/w
    h = jn_zeros(0,5)[0]* w / 4.0
    params = {'h0':np.pi/32, 'h':h,'omega':w, 'N':N} 
    H0, H1 = get_hamiltonians(N)
    H = [H0,[H1,drive]]
    en, sts = sx.eigenstates() 
    rho0 = sts[0]
    times = np.linspace(0,75*T, 500)
    hbar = []
    out = mesolve(H, rho0, times, [], [], args = params, options=opts)
    psi_ts = out.states
    
    for i,t in enumerate(times):
        Ht = H0 + H1 * drive(t, params)
        hbar.append(expect(Qobj(Ht),psi_ts[i]))
    hbar = (np.array(hbar)/N).reshape(len(times),1)
    return hbar.real, (times/(2*np.pi/w)).reshape(len(times),1)

if __name__ == '__main__':
    nprocs = 8
    N = 20
    o1 = np.linspace(0.29, 1.58, 5)
    o2 = np.linspace(1.6, 2.5, 6)
    o3 = np.linspace(2.8, 4.38, 6)
    o4 = np.linspace(4.5, 9.4, 5)
    o5 = np.linspace(9.5, 20.0, 5)
    o6 = np.linspace(20.1, 50.0, 9)
    
    o7 = np.linspace(0.1, 2, 15)
    o8 = np.linspace(1.6, 2.5, 15)
    o9 = np.linspace(2.8, 4.38, 15)
    
    o10 = np.linspace(0.05, 2, 10)


    omega_vals = np.concatenate((o1, o2, o3, o4, o5, o6))
    #omega_vals = np.concatenate((o7,o8,o9))
    #omega_vals = o10
    
    p = Pool(processes = nprocs)
    print("running for TSS spin N=",N, 'nprocs=',nprocs," !")
    opts = Options(nsteps=1e5, num_cpus=nprocs, openmp_threads=1, atol=1e-12, rtol=1e-14)
    data = np.array(p.starmap(sys_evolve,tqdm([(w,N, opts) for w in omega_vals])))
```

```python
hbaravg =[]
for i,w in enumerate(omega_vals):
    hbar = data[i][0][:,0]
    times = data[i][1][:,0]
    
    hbaravg.append(np.average(hbar))
plt.plot(omega_vals,hbaravg)
plt.xlabel(r'$\omega$')
plt.ylabel(r"$\langle H(t) \rangle$")
plt.xlim(-1,10)
plt.axhline(y=1.0,color='r', linestyle='--', alpha=0.5)
plt.show()
```

## Floquet modes $<S_x> $

\begin{equation}
U_{nm} = \sum_\alpha exp\Big(i\frac{\Omega_\alpha t}{\hbar}\Big) \langle n\vert \alpha\rangle\langle\alpha\vert m\rangle
\end{equation}
The Floquet mode- magenetisation is defined as
\begin{equation}
\langle S_x \rangle_{Floquet} = U_{nm}^\dagger S_x U_{nm}
\end{equation}

where $\vert \alpha \rangle$ and $\Omega_\alpha$ is $\alpha$-th Floquet modes and quasienergies. 

```python
%%time

from qutip import mesolve, basis, jmat, Options, expect, Qobj,floquet_modes, propagator
from multiprocessing import Pool
from tqdm import tqdm
from scipy.special import jn_zeros
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

plt.rcParams.update({"figure.figsize": (12,7),"text.usetex": True,
    "font.family": "sans-serif","font.size": 18,"font.sans-serif": ["Computer Modern"]})

def drive(t, args):
    w = args['omega']
    h = args['h']
    h0 = args['h0']
    return h * np.sin(w*t) + h0

def propagator_evolve(N,w,params, opts):
    T = 2 * np.pi/w
    H0, H1 = get_hamiltonians(N)
    H = [H0,[H1,drive]]
    U = propagator(H, T, [], params, parallel=False, progressbar=False, options=opts)
    U = np.array(U.full())
    return U

def get_hamiltonians(N):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    kn =  2.0/(N-1)                                     
    H0 = kn * sz **2 
    H1 = 2 * sx
    return H0,H1

def sys_evolve(w,N,opts):
    sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z")
    T = 2 * np.pi/w
    h = jn_zeros(0,5)[0]* w / 4.0
    params = {'h0':np.pi/32, 'h':h,'omega':w, 'N':N} 
    H0, H1 = get_hamiltonians(N)
    H = [H0,[H1,drive]]  
    f_modes_0, f_energies = floquet_modes(H, T, params, options= opts)
    U = propagator_evolve(N,w,params, opts)
    evalOmega, evecOmega = la.eig(U)
    return [f_modes_0, f_energies], evalOmega

if __name__ == '__main__':
    nprocs = 12
    
    o1 = np.linspace(0.29, 0.9, 10)
    o2 = np.linspace(1.0, 2.5, 10)
    o3 = np.linspace(2.7, 4.38, 12)
    o4 = np.linspace(4.5, 9.4, 12)
    o5 = np.linspace(9.5, 20.0, 12)
    o6 = np.linspace(20.1, 40.0, 4)
    
    o7 = np.linspace(0.1, 2, 15)
    o8 = np.linspace(1.0, 2.5, 15)
    o9 = np.linspace(2.8, 4.38, 15)
    
    o10 = np.linspace(.5, 5, 36)


    omega_vals = np.concatenate((o1,o2,o3, o4,o5, o6))
    #omega_vals = np.concatenate((o7,o8,o9))
    omega_vals = o10
    
    Ns = [20, 50, 80]
    nstp = [1e5, 1e5, 1e6]
    
    for n, N in enumerate(Ns):
        p = Pool(processes = nprocs)
        print("running for TSS spin N=",N, 'nprocs=',nprocs," !")
        opts = Options(nsteps=nstp[n], num_cpus=nprocs, openmp_threads=1, atol=1e-12, rtol=1e-14)
        data = np.array(p.starmap(sys_evolve,tqdm([(w,N, opts) for w in omega_vals]))) 
        
        sx,sy,sz = jmat(N,"x"),jmat(N,"y"),jmat(N,"z") 
        sxuavg = []
        
        for iw in range(len(omega_vals)):
            fl_modes = data[iw][0][0]
            # change the basis for floquet modes from sz to sx due to zzx model
            en, st = sx.eigenstates()
            f_modes = [Qobj(np.array([[f_m.overlap(sts) for sts in st ] for f_m in fl_modes])[i]) for i in range(2*N+1)]
            
            unm = Qobj(np.sum([data[iw][1][i] * np.array(Qobj(f_modes[i]).proj()) for i in range(2*N+1)], axis=0))
            #sxu = unm.dag() * sx * unm
            sxu = sx
            sxuavg.append(np.average([expect(sxu, f_modes[i]) for i in range(2*N+1)]))
            
        plt.plot(omega_vals, sxuavg, label = fr"$N={N:3d}$")
    plt.legend(frameon=False)
    plt.xlabel(r"$\omega$")
    plt.ylabel(r"$\langle S_{x_{{_{{Flq}}}}} \rangle$")
    plt.show()
```
